# Предобработка данных

## ferma_main

Измененные столбцы:

1. id — **id**
2. удой, кг — **milk_yield**
3. ЭКЕ — **fodder_energy_unit** 
4. сырой протеин, г — **crude_protein**
5. СПО — **sugar_protein_ratio**  
6. порода — **breed**  
7. тип пастбища — **pasture**
8. порода папы_быка — **dad_breed**  
9. жирность, % — **fat_content** 
10. белок, % — **protein_content**
11. вкус молока — **milk_taste**  
12. возраст — **age_more_than_2_years** (булево значение - больше ли двух лет)

```python
ferma_main.columns = [
    'id',
    'milk_yield',
    'fodder_energy_unit',
    'crude_protein',
    'sugar_protein_ratio',
    'breed',
    'pasture',
    'dad_breed',
    'fat_content',
    'protein_content',
    'milk_taste',
    'age'
]
```

Проверим строки-дубликаты

```python
ferma_main.duplicated().sum()
```

```python
ferma_main = ferma_main.drop_duplicates()
```

Проверим уникальные значения в категориальных данных

```python
ferma_main['milk_taste'].unique(), ferma_main['age'].unique()
```

```python
ferma_main['breed'].unique(), ferma_main['pasture'].unique(), ferma_main['dad_breed'].unique()
```

Некоторые вещественные значения содержат `,` вместо `.`. В столбце с типом пастбища есть значения, которые стоит объединить в одно: `Равнинное`, `Равнинные`.

```python
ferma_main = ferma_main.drop_duplicates()
ferma_main['pasture'] = ferma_main['pasture'].str.replace('Равнинные', 'равнинное')
ferma_main['pasture'] = ferma_main['pasture'].str.replace('Равнинное', 'равнинное')
ferma_main['pasture'] = ferma_main['pasture'].str.replace('Холмистое', 'холмистое')
ferma_main['dad_breed'] = ferma_main['dad_breed'].str.replace('Айдиалл', 'Айдиал')
```

```python
ferma_main.isna().sum()
```

```python
y_proba = clf.predict_proba(X_test)[:,1]

data = pd.DataFrame(
    zip(y_test, y_proba),
    columns = ['y_valid', 'y_proba']
).sort_values(by='y_proba',ascending=False)

print(data.head(20))
```

```python
thresholds = [round(i,2) for i in np.linspace(0.1,1,num = 10,endpoint=False)]
columns = []

for i in thresholds:
    columns.append('y_pred_' + str(i))
    data['y_pred_' + str(i)] = data['y_proba'].apply(lambda x: 1 if x >= i else 0 )
print(data.sample(5))
```
