# Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекс Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Необходимо научиться определять рыночную стоимость объектов недвижимости. Для этого нужно провести исследовательский анализ данных и установить параметры, влияющие на цену объектов. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность.

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта и других объектов — эти данные автоматически получены из геосервисов. Количество парков и водоёмов также заполняется без участия пользователя.

### Предобработка данных

Переименуем некоторые столбцы

```python
data.rename(columns={
    'cityCenters_nearest' : 'city_centers_nearest', 
    'parks_around3000' : 'parks_around_3000', 
    'ponds_around3000' : 'ponds_around_3000', 
}, inplace=True)
```

#### Выведем количество пропущенных значений для каждого столбца

```python
data.isna().sum()
```

В столбце `balcony` пропуски заменим значениями 0 (так как если продавец не указал наличие того, что повышает стоимость квартиры, то, скорее всего, в его квартире этого просто нет). Аналогично поступим со столбцами `is_apartment`, `parks_around_3000`, `ponds_around_3000`. Остальные пропуски нельзя заполнить логически, так как у каждой квартиры свои параметры (площадь, расстояние до определенных мест, сколько дней было размещено объявление)

```python
data['balcony'] = data['balcony'].fillna(0)
data['is_apartment'] = data['is_apartment'].fillna('False')
data['parks_around_3000'] = data['parks_around_3000'].fillna(0)
data['ponds_around_3000'] = data['ponds_around_3000'].fillna(0)

for t in data['locality_name'].unique():
    if data.loc[(data['locality_name'] == t), 'ceiling_height'].count() > 0:
        data.loc[(data['locality_name'] == t) & (data['ceiling_height'].isna()), 'ceiling_height'] = \
        data.loc[(data['locality_name'] == t), 'ceiling_height'].median() 
```

Заменим типы значений в определенных столбцах. В столбце `is_apartment` содержатся исключительно булевые значения, поэтому тип изменен на bool. В столбцах `balcony`, `parks_around_3000`, `ponds_around_3000` содержатся количественные значения, поэтому рационально заменить их на int. В столбцах `floors_total`, `days_exposition` не получится изменить тип, так как там есть пропущенные значения, которые нельзя как-то изменить (а тип NaN является вещественным значением)

```python
data['is_apartment'] = data['is_apartment'].astype(bool)
data['balcony'] = data['balcony'].astype(int)
data['ponds_around_3000'] = data['ponds_around_3000'].astype(int)
data['parks_around_3000'] = data['parks_around_3000'].astype(int)
```

#### Изучим уникальные значения в столбце с названиями населенных пунктов.

```python
data['locality_name'].unique()
```

Изменим все буквы ё на е, чтобы исключить слово посёлок, а также разделим `locality_name` на 2 столбца так, что в `locality_type` будет наименование типа местности, а в `locality_clean_name` его название без типа. Будет использовано регулярное выражение для того, чтобы разделить строку там, где есть заглавная буква кириллицей.

```python
data['locality_name'] = (
    data['locality_name']
    .str.replace('ё', 'е')
)
data.loc[:, ['locality_type', 'locality_clean_name']] = \
data['locality_name'].str.split(pat='(?=[A-Я])', n=1, expand=True).to_numpy()
```

### Добавление в таблицу новых столбцов

Цена одного квадратного метра образуется делением стоимости объекта на его общую площадь, а затем округлением до двух знаков после запятой). Год, месяц и день недели публикации образуется при помощи перевода даты в DatetimeIndex и извлечением определенного свойства данного объекта.

```python
data['one_metre_price'] = round(data['last_price'] / data['total_area'], 2)
data['weekday'] = pd.DatetimeIndex(data['first_day_exposition']).weekday
data['month'] = pd.DatetimeIndex(data['first_day_exposition']).month
data['year'] = pd.DatetimeIndex(data['first_day_exposition']).year
```

Функция `get_floor_type` понадобится для определения типа этажа. 

```python
def get_floor_type(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
```

```python
data['floor_type'] = data.apply(get_floor_type, axis=1)
```

```python
data['city_centers_nearest_km'] = round(data['city_centers_nearest'] / 1000)
```

Рассмотрим значения от 2 до 5 для понимания наиболее распространенных значений среди наиболее реальных

```python
data['ceiling_height'].hist(bins=100, range=(2, 5));
```

Значения от 24 до 50 вполне могут быть значениями от 2.4 до 5.0. А потолки до 2.4 м, а также от 5 метров не кажутся реалистичными. Заменим их на соответствующие значения. 

```python
def change_ceiling_height(ceiling_height):
    if 24 <= ceiling_height <= 50:
        return round(ceiling_height / 10)
    else:
        return ceiling_height
    
    
data['ceiling_height'] = data['ceiling_height'].apply(change_ceiling_height)
```

```python
data = data.loc[(data['ceiling_height'] >= 2.4) & (data['ceiling_height'] <= 5)]
```

#### Посчитаем среднюю цену одного квадратного метра в 10 населённых пунктов с наибольшим числом объявлений

Создадим таблицу из 10 населённых пунктов с наибольшим числом объявлений

```python
top_ten_cities = data.pivot_table(index='locality_name', values='last_price', aggfunc='count').sort_values(
    by='last_price', ascending=False).head(10)
top_ten_cities.columns = ['count']
top_ten_cities
```

Посчитаем среднюю стоимость квадратного метра для каждого города из топ-10.

```python
def calculate_mean_price(city):
    return data.loc[data['locality_name'] == city, 'one_metre_price'].mean()


top_ten_cities['one_metre_price'] = top_ten_cities.index.to_series().apply(
    calculate_mean_price)
top_ten_cities
```

**Таким образом, в Санкт-Петербурге (112929) самая дорогая стоимость квадратного метра, а в Выборге (58238) - самая дешевая.**

#### Выделим квартиры в Санкт-Петербурге с помощью столбца locality_name и вычислим их среднюю стоимость на разном удалении от центра 

```python
data[data['locality_name'] == 'Санкт-Петербург'].groupby('city_centers_nearest_km')['last_price'].mean()\
    .plot(x='city_centers_nearest_km', y='last_price', style='-o');
```

**Квартиры в Санкт-Петербурге при близости к центру увеличиваются в стоимости, но есть исключения.**
