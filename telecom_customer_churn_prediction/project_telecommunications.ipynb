# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

```python
RANDOM_STATE = 50525
TEST_SIZE = 0.25
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(
        [
            'customer_id',
            'agreement_termination'
        ],
        axis=1
    ),
    df['agreement_termination'],
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=df['agreement_termination']
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# –ü—Ä–∏–∑–Ω–∞–∫–∏ –¥–ª—è One-Hot Encoding (OHE)
ohe_columns = [
    'type',
    'paperless_billing',
    'payment_method',
    'senior_citizen',
    'partner',
    'dependents',
    'internet_service',
    'online_security',
    'online_backup',
    'device_protection',
    'tech_support',
    'streaming_tv',
    'streaming_movies',
    'multiple_lines'
]

# –ß–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
num_columns = [
    'monthly_charges',
    'total_charges',
    'services_count',
    'begin_date'
]
```

```python
# SimpleImputer + OHE
ohe_pip = Pipeline(
    [
        (
            'simpleImputer_ohe',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe',
            OneHotEncoder(drop='first', handle_unknown='ignore',
                          sparse_output=False)
        )
    ]
)
```

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pip, ohe_columns),
        ('num', MinMaxScaler(), num_columns)
    ],
    remainder='passthrough'
)
```

**–í—ã–≤–æ–¥—ã:**
- –î–∞–Ω–Ω—ã–µ —Ä–∞–∑–¥–µ–ª–µ–Ω—ã –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏ (75% –¥–ª—è —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫–∏, 25% –¥–ª—è —Ç–µ—Å—Ç–∞).
- –¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è ‚Äî `agreement_termination`, –æ—Å—Ç–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ ‚Äî –∫–∞–∫ —Ñ–∏—á–∏.
- –î–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–Ω–∞–ø—Ä–∏–º–µ—Ä, `type`, `payment_method`) –ø—Ä–∏–º–µ–Ω—ë–Ω –º–µ—Ç–æ–¥ –∑–∞–º–µ–Ω—ã –ø—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–∞ –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è, –∞ –∑–∞—Ç–µ–º —ç—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–∏ –ø—Ä–µ–æ–±—Ä–∞–∑—É—é—Ç—Å—è –≤ –±–∏–Ω–∞—Ä–Ω—ã–µ —Å –ø–æ–º–æ—â—å—é `OneHotEncoder`.
- –î–ª—è —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è `MinMaxScaler` –¥–ª—è –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏–∏ –∑–Ω–∞—á–µ–Ω–∏–π –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç 0 –¥–æ 
- –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω `ColumnTransformer`, –∫–æ—Ç–æ—Ä—ã–π –æ–±—ä–µ–¥–∏–Ω—è–µ—Ç –æ–±—Ä–∞–±–æ—Ç–∫—É –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –∏ —á–∏—Å–ª–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –≤ –æ–¥–Ω–æ–º —à–∞–≥–µ.

# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è

–°–æ–∑–¥–∞–¥–∏–º –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω.

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

## LogisticRegression

```python
pipe_logreg = Pipeline([
    ('preprocessor', data_preprocessor),
    ('model', LogisticRegression(
        random_state=RANDOM_STATE,
        solver='liblinear',
        penalty='l1'
    ))
])

param_grid_logreg = {
    'model__C': range(1,5),
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
}

search_logreg = RandomizedSearchCV(
    pipe_logreg,
    param_distributions=param_grid_logreg,
    scoring='roc_auc',
    cv=3,
    random_state=RANDOM_STATE,
    n_jobs=-1
)

search_logreg.fit(X_train, y_train)
search_logreg.best_score_
```

## DecisionTreeClassifier

```python
pipe_tree = Pipeline([
    ('preprocessor', data_preprocessor),
    ('model', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

param_tree = {
    'model__max_depth': range(2, 10),
    'model__max_features': range(2, X_train.shape[1]),
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
}

search_tree = RandomizedSearchCV(
    pipe_tree,
    param_distributions=param_tree,
    cv=3,
    scoring='roc_auc',
    n_jobs=-1,
    random_state=RANDOM_STATE
)

search_tree.fit(X_train, y_train)

search_tree.best_score_
```

## KNeighborsClassifier

```python
pipe_knn = Pipeline([
    ('preprocessor', data_preprocessor),
    ('model', KNeighborsClassifier())
])

param_knn = {
    'model__n_neighbors': range(2, 20),
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
}

search_knn = RandomizedSearchCV(
    pipe_knn,
    param_distributions=param_knn,
    cv=3,
    scoring='roc_auc',
    n_jobs=-1,
    random_state=RANDOM_STATE
)

search_knn.fit(X_train, y_train)

search_knn.best_score_
```

Recall = 0.69 ‚Äî –º–æ–¥–µ–ª—å –≤—ã—è–≤–ª—è–µ—Ç –æ–∫–æ–ª–æ 69% –≤—Å–µ—Ö –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤, –Ω–æ –æ—Å—Ç–∞–≤–ª—è–µ—Ç –Ω–µ–º–∞–ª—É—é —á–∞—Å—Ç—å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö (–ª–æ–∂–Ω—ã—Ö –æ—Ç—Ä–∏—Ü–∞–Ω–∏–π).

```python
best_model = search_logreg.best_estimator_
coefficients = best_model.named_steps['model'].coef_[0]
ohe_columns_transformed = best_model.named_steps['preprocessor'].transformers_[0][1].get_feature_names_out(ohe_columns)
feature_names = np.concatenate([ohe_columns_transformed, num_columns])
importance_df = pd.DataFrame({
    '–ü—Ä–∏–∑–Ω–∞–∫': feature_names,
    '–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç': coefficients
}).sort_values(by='–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç', ascending=False)
importance_df
```
